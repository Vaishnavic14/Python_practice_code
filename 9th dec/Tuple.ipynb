# Loop through a tuple

```python
mytuple = ['one' , 'two' , 'three' , 'four' , 'five' , 'six' , 'seven' , 'eight']
mytuple
```

```python
for i in mytuple:
   print(i)
```

```python
for i in enumerate(mytuple):
   print(i)
```
