##  Global & Local Argument

```python
a = 10  #-- globla variable 

def something():
    b = 15 #local variable
    
    print('in function',b)    
    print('out function',a) 
```

```python
a = 10  

def something():
    b = 15 
    print('in function',b)    
    
print('out function',a)  
```

```python
a = 10  

def something():
    b = 15 
    print('in function',b)    
    
print('out function',a) 

something()
```

```python
x = 10  # Global variable

def update_x():
    global x  # Declare that we are using the global variable x
    x += 5    # Modify the global variable

update_x()
print(x)  # Output: 15
```

```python
x = 10  # Global variable

def update_x():
    globals()['x'] += 5  # Access and modify the global variable using the dictionary

update_x()
print(x)  # Output: 15
```
